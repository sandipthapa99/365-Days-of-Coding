```python
import matplotlib.pyplot as plt

class1 = [70,65,82,55,78,80,55,66,87,48,54]
class2 = [32,46,35,30,45,54,55,80,29,36,38]
class_range = [10,20,30,40,40,60,70,80,90,100,110]

fig = plt.figure()
axis = fig.add_axes([0,0,1,1])
axis.scatter(class_range, class1, color='b')
axis.scatter(class_range, class2, color='r')
axis.set_xlabel('Class Range')
axis.set_ylabel('Class Score')
axis.set_title('Scatter Plot')
axis.legend(["Class1","Class2"])
plt.show()
```
